```python
import csv

#load polarity dictionary
file = open('polarity.csv','r',encoding='UTF-8')
reader = csv.reader(file)

#get header
header = reader.__next__()

#make dictionary
pos_dic = [[],[],[],[],[],[],[],[],[]]  #긍정
neg_dic = [[],[],[],[],[],[],[],[],[]]  #부정
comp_dic = [[],[],[],[],[],[],[],[],[]] #복합
neut_dic = [[],[],[],[],[],[],[],[],[]] #중립

for line in reader:
    line[0] = line[0].split(';') #split ngram
    maxValue = line[7]
    
    if maxValue == 'POS':
        for i in range(len(header)):
            pos_dic[i].append(line[i])
    elif maxValue == 'NEG':
        for i in range(len(header)):
            neg_dic[i].append(line[i])
    elif maxValue == 'COMP':
        for i in range(len(header)):
            comp_dic[i].append(line[i])
    else:
        for i in range(len(header)):
            neut_dic[i].append(line[i])
```
