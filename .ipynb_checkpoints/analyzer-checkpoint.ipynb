```python
import csv
from konlpy.tag import Kkma

#load polarity dictionary
file = open('./dic/polarity.csv','r',encoding='UTF-8')
reader = csv.reader(file)

#get header
header = reader.__next__()

#make dictionary
pos_dic = [[],[],[],[],[],[],[],[],[]]  #긍정
neg_dic = [[],[],[],[],[],[],[],[],[]]  #부정
comp_dic = [[],[],[],[],[],[],[],[],[]] #복합
neut_dic = [[],[],[],[],[],[],[],[],[]] #중립

for line in reader:
    line[0] = line[0].split(';') #split ngram
    maxValue = line[7]
    
    if maxValue == 'POS':
        for i in range(len(header)):
            pos_dic[i].append(line[i])
    elif maxValue == 'NEG':
        for i in range(len(header)):
            neg_dic[i].append(line[i])
    elif maxValue == 'COMP':
        for i in range(len(header)):
            comp_dic[i].append(line[i])
    else:
        for i in range(len(header)):
            neut_dic[i].append(line[i])
```

```python
def getBigram(data):
    ret = []
    for i in range(len(data)):
        if i==len(data)-1:
            break
        p = []
        p.append(data[i])
        p.append(data[i+1])
        ret.append(p)
    return ret

def getTrigram(data):
    ret = []
    for i in range(2,len(data)):
        p = []
        p.append(data[i-2])
        p.append(data[i-1])
        p.append(data[i])
        ret.append(p)
    return ret

def findPOS(unigram,bigram,trigram):
    pos_val = 0
    for uni in unigram:
        if uni in pos_dic[0]:
            mvPos=pos_dic[0].index(uni) #max value position
            mvProp=pos_dic[8][mvPos] #max value prop
            #print('POS',uni,'POSITIVE',mvProp) 
            pos_val += float(mvProp)
    for bi in bigram:
        if bi in pos_dic[0]:
            mvPos=pos_dic[0].index(bi) #max value position
            mvProp=pos_dic[8][mvPos] #max value prop
            #print('POS',bi,'POSITIVE',mvProp)
            pos_val += float(mvProp)
    for tri in trigram:
        if tri in pos_dic[0]:
            mvPos=pos_dic[0].index(tri) #max value position
            mvProp=pos_dic[8][mvPos] #max value prop
            #print('POS',tri,'POSITIVE',mvProp)
            pos_val += float(mvProp)
            
    return pos_val

def findNEG(unigram,bigram,trigram):
    neg_val = 0
    for uni in unigram:
        if uni in neg_dic[0]:
            mvPos=neg_dic[0].index(uni) #max value position
            mvProp=neg_dic[8][mvPos] #max value prop
            #print('NEG',uni,'NEGATIVE',mvProp) 
            neg_val += float(mvProp)
    for bi in bigram:
        if bi in neg_dic[0]:
            mvPos=neg_dic[0].index(bi) #max value position
            mvProp=neg_dic[8][mvPos] #max value prop
            #print('NEG',bi,'NEGATIVE',mvProp) 
            neg_val += float(mvProp)
    for tri in trigram:
        if tri in neg_dic[0]:
            mvPos=neg_dic[0].index(tri) #max value position
            mvProp=neg_dic[8][mvPos] #max value prop
            #print('NEG',tri,'NEGATIVE',mvProp) 
            neg_val += float(mvProp)
            
    return neg_val
    
    
```

```python
# kkma = Kkma()
# print(kkma.pos('너는 몇살인데 나한테 반말질이야?'))

#file open
corpus = open('./corpus/BTAA0011.txt','r',encoding='UTF-8')

positive_value = 0
negative_value = 0
sentence=''
#ngram
for line in corpus:
    if(line!='\n'):
        sentence +=line.split('\n')[0].split('\t')[0]+' '
        unigram = line.split('\n')[0].split('\t')[1].split(' + ')
        bigram = getBigram(unigram)
        trigram = getTrigram(unigram)
        positive_value += findPOS(unigram,bigram,trigram)
        negative_value += findNEG(unigram,bigram,trigram)
        
    else:
        
        print(sentence)
        print('-----------------------------------------')
        print('해당 문장의 긍정점수 : ',positive_value)
        print('해당 문장의 부정점수 : ',negative_value)
        sum = positive_value+negative_value
        if(sum!=0):
            print('긍정 부정 퍼센티지 : ',round(positive_value/sum*100,2),'%')
            sentence=''
            positive_value = 0
            negative_value = 0
        print('\n\n')
```
